```python
#-------------------------------------------------First Figure-----------------------------------------------------------
import matplotlib.pyplot as plt
import networkx as nx

# Create directed graph
G = nx.DiGraph()

# Point location
positions = {
    'P1': (0, 2),
    'A1': (1, 0),
    'B': (-1, 0),
    'K': (0,-2),
    'P2': (2, -2),
    'D': (1, -4),
    'A2': (3, -4),
    'J':(4,-6),
    'M':(2,-6)
}

#draw connection between edges
edges = [
    ('P1', 'A1', 'In'),
    ('A1', 'K', 'Low'),
    ('A1', 'P2', 'High'),
    ('P2', 'D', 'Out'),
    ('P2', 'A2', 'In'),
    ('P1', 'B', 'Out'),
    ('A2', 'J', 'High'),
    ('A2','M','Low'),
]

# Add edges to the graph with labels(listcompre)
for edge in edges:
    G.add_edge(edge[0], edge[1], label=edge[2])

node_labels = {node: node for node in G.nodes()}


# Define payoffs and node labels
payoffs = {
    'B': '(5,5)',
    'D': '(0,10)',
    'K': '(0,14)',
    'J':'(12,10)',
    'M':'(0,10)',
}
node_labels = {node: node for node in G.nodes()}

del node_labels['D']
del node_labels['K']
del node_labels['M']
del node_labels['B']
del node_labels['J']
#drawing
specific_colors = {'P2': 'Gray', 'A2': 'Gray'}
# Default color for other nodes
default_color = 'Pink'
specific_sizes = {'D':0,'K':0,'M':0,'B':0,'J':0}
default_nodesize = 2200

# Generate a list of colors in the order of G.nodes()
node_colors = [specific_colors.get(node, default_color) for node in G.nodes()]
# Put certain node size to 0 to actively remove the circle
node_sizes =[specific_sizes.get(node,default_nodesize) for node in G.nodes()]
nx.draw(G, pos=positions, with_labels=True, node_size=node_sizes, node_color=node_colors, labels=node_labels)

# Draw edge labels
edge_labels = {(u, v): d['label'] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos=positions, edge_labels=edge_labels, label_pos=0.5)

                   
for node, payoff in payoffs.items():
    x, y = positions[node]
    plt.text(x, y-0.4, payoff, fontsize=13, ha='center')
plt.text(2.4,-2,'<----Strategic Uncertainty')
plt.text(3.4,-4,'<----Strategic Uncertainty')
plt.axis('off')  # Turn off the axis
plt.title('Game Tree')
plt.show()  

```

```python
#------------------------------------------------------Second Figure---------------------------------------------------------

import matplotlib.pyplot as plt
import networkx as nx

# Create directed graph
G = nx.DiGraph()

# Point locations
positions = {
    'P1': (0, 2),
    'A1': (1, 1),
    'P2 ': (-2.5, 0),
    'P2  ': (0, 0),
    'P2': (2, 0),
    'D': (1, -1),
    'A2': (3.25, -1.5),
    'J': (4.5, -3),
    'M': (2, -3),
    'A2_1': (-1.25, -1),
    'A2 ': (0.5, -1.5),
    'A2  ': (-2, -1.5),
    'A2_4': (-4, -1),
    'C1': (1.5, -3),
    'C2': (-0.75, -3),
    'C3': (-1.25, -3),
    'C4': (-3.5, -3)
}

# Add nodes to the graph
G.add_nodes_from(positions.keys())

# Add edges to the graph
G.add_edge('P1', 'A1', label='In', label_pos = 0.5)
G.add_edge('P1', 'P2 ', label='Out')
G.add_edge('A1', 'P2  ', label='Low')
G.add_edge('A1', 'P2', label='High')
G.add_edge('P2', 'D', label='Out')
G.add_edge('P2', 'A2', label='In')
G.add_edge('A2', 'J', label='High')
G.add_edge('A2', 'M', label='Low')
G.add_edge('P2  ', 'A2 ', label='In')
G.add_edge('P2  ', 'A2_1', label='Out')
G.add_edge('P2 ', 'A2  ', label="In")
G.add_edge('P2 ', 'A2_4', label="Out")
G.add_edge('A2 ', 'C1', label="High")
G.add_edge('A2 ', 'C2', label="Low")
G.add_edge('A2  ', 'C3', label="High")
G.add_edge('A2  ', 'C4', label="Low")

# Nodes to delete labels
delete_node_labels = ['A2_4', 'A2_1', 'D', 'C4', 'C3', 'C2', 'C1', 'M', 'J']

# Create a label dictionary for nodes you want to label
node_labels = {node: node for node in G.nodes() if node not in delete_node_labels}

# Color and size settings
specific_colors = {'P1': 'royalblue', 'A1': 'darkviolet', 'P2 ': '#90EE90', 'P2  ': '#90EE90', 'P2': '#90EE90'}
default_color = '#FF4500'

node_colors = [specific_colors.get(node, default_color) for node in G.nodes()]

specific_sizes = {'D':0,'K':0,'M':0,'B':0,'J':0,'C2':0,'C1':0,'C3':0,'C4':0,'A2_1':0,'A2_4':0}
default_nodesize = 520
node_sizes =[specific_sizes.get(node,default_nodesize) for node in G.nodes()]

payoff_color = {'P1': 'blue', 'A1': 'purple', 'P2 ': 'green', 'P2  ': 'green', 'P2': 'green'}
node_colors = [specific_colors.get(node, default_color) for node in G.nodes()]

payoffs = {'D': [0,10,5,5], 'M': [0,10,0,14], 'J': [12,10,12,10], 'C2': [0,14,0,14], 'C1': [0,14,0,10], 'C3': [5,5,0,10], 'C4': [5,5,0,14], 'A2_1': [0,14,5,5],'A2_4':[5,5,5,5]}

#order of players for payoffs
player_order = ['P1', 'A1', 'P2', 'A2']

# Draw the network
nx.draw(G, pos=positions, labels=node_labels, node_size=node_sizes, node_color=node_colors, font_size=11, font_weight='bold', arrowsize=12, font_family='Times New Roman')

# Draw edge labels
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos=positions, edge_labels=edge_labels, font_family='Times New Roman')

for node, values in payoffs.items():
    x, y = positions[node]
    for i, value in enumerate(values):
        plt.text(x-0.2, y-0.18*i, f'{value}', color=payoff_color.get(player_order[i],'red'), fontsize=10, ha='center',va='bottom',fontname='Times New Roman')
    

plt.text(-2.5,0.05,'     _ _ _ _ _ _ _ _ _ _ _ _         _ _ _ _ _ _ _ _ _       ',fontname='Times New Roman')
plt.text(-1.9, -1.555,'   _ _  _ _ _ _ _ _ _ _ _ _        _ _ _ _ _ _ _ _ _ _ _ _ _ _        ',fontname='Times New Roman')           
#x = 6
#y = 2
#plt.text(x,y, 'Payoff for\n   P1\n   A1\n   P2\n   A2')

import matplotlib.pyplot as plt

# Your starting coordinates
x = 6
y = 2

# The offset for each line of text
offset = 0.2  # Adjust as needed for the gap between "payoff for"

# Plot each line of text
plt.text(x-4.45, y, 'Payoff for', color='black', ha='left', va='center',fontname='Times New Roman',fontsize=11 )
plt.text(x-4.1, y - offset, 'P1', color='blue', ha='left', va='center',fontname='Times New Roman',fontsize=11)
plt.text(x-4.1, y - 2 * offset, 'A1', color='purple', ha='left', va='center',fontname='Times New Roman',fontsize=11)
plt.text(x-4.1, y - 3 * offset, 'P2', color='green', ha='left', va='center',fontname='Times New Roman',fontsize=11)
plt.text(x-4.1, y - 4 * offset, 'A2', color='red', ha='left', va='center',fontname='Times New Roman',fontsize=11)

plt.savefig('ProfCason_graphfinal.png', format='png', dpi=300, bbox_inches='tight')
plt.show()

plt.axis('off')  # Turn off the axis
plt.show()
```
